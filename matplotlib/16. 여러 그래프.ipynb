# 16. 여러 그래프

```python
import matplotlib.pyplot as plt
import matplotlib

# 한글을 사용하기 위한
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 맑은 고딕, Window
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd

df = pd.read_excel('../Pandas/score.xlsx')
df
```

```python
fig, axs = plt.subplots(2, 2, figsize = (15, 8)) # 2 * 2에 해당하는 plot들을 생성
plt.suptitle('여러 그래프 넣기')

# 첫 번째 그래프 - 막대 그래프(세로)
axs[0, 0].bar(df['이름'], df['국어'], label = '국어 점수') # 데이터 설정
axs[0, 0].set_title('첫 번째 그래프') # 제목
axs[0, 0].legend() # 범례
axs[0, 0].set(xlabel = '이름', ylabel = '점수') # x, y축 label
axs[0, 0].set_facecolor('lightyellow') # 전면 색
axs[0, 0].grid(ls = '--', linewidth = 0.5)

# 두 번째 그래프 - 꺾은 선 그래프
axs[0, 1].plot(df['이름'], df['영어'], label = '영어 점수') # 데이터 설정
axs[0, 1].plot(df['이름'], df['수학'], label = '수학 점수') 
axs[0, 1].set_title('두 번째 그래프') # 제목
axs[0, 1].legend() # 범례
axs[0, 1].set(xlabel = '이름', ylabel = '점수') # x, y축 label
axs[0, 1].set_facecolor('pink') # 전면 색
axs[0, 1].grid(ls = '--', linewidth = 0.5, color = 'black')

# 세 번째 그래프 - 막대 그래프(가로)
axs[1, 0].barh(df['이름'], df['키'], label = '영어 점수') # 데이터 설정
axs[1, 0].set_title('세 번째 그래프') # 제목
axs[1, 0].legend() # 범례
axs[1, 0].set(xlabel = '이름', ylabel = '키') # x, y축 label
axs[1, 0].set_facecolor('purple') # 전면 색
axs[1, 0].grid(ls = '--', linewidth = 0.5, color = 'white')

# 네 번째 그래프 - 막대 그래프(가로)
axs[1, 1].plot(df['이름'], df['사회'], label = '사회 점수') # 데이터 설정
axs[1, 1].set_title('네 번째 그래프') # 제목
axs[1, 1].legend() # 범례
axs[1, 1].set(xlabel = '이름', ylabel = '점수') # x, y축 label
# axs[1, 1].set_facecolor('purple') # 전면 색
axs[1, 1].grid(ls = '--', linewidth = 0.5, color = 'white')
```
