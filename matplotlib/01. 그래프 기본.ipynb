# Matplotlib
다양한 형태의 그래프를 통해서 데이터를 시각화할 수 있는 라이브러리

```python
import matplotlib.pyplot as plt
```

# 1. 그래프 기본

```python
x = [1, 2, 3]
y = [2, 4, 8]
plt.plot(x, y)
```

```python
print(plt.plot(x, y))
```

```python
plt.plot(x, y)
plt.show() # [<matplotlib.lines.Line2D at 0x26d943cf340>] 없이 그래프만 보고싶을 때
```

## Title 설정

```python
plt.plot(x, y)
plt.title('Line Graph')
```

## 한글 폰트 설정

```python
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 맑은 고딕, Window
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False #  한글 폰트 사용시, 마이너스 글자가 깨지는 것을 방지

# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
```

```python
import matplotlib.font_manager as fm
fm.fontManager.ttflist # 사용 가능한 폰트 확인
[f.name for f in fm.fontManager.ttflist]
```

```python
matplotlib.rcParams['font.family'] = 'HYGungSo-Bold' # 궁서체
plt.plot(x, y)
plt.title('꺾은선 그래프')
```
