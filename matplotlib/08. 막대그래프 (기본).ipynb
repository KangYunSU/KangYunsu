# 8. 막대 그래프 (기본)

```python
import matplotlib.pyplot as plt
import matplotlib

# 한글을 사용하기 위한
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 맑은 고딕, Window
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]

plt.bar(labels, values)
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]

plt.bar(labels, values, color = 'red')
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]
colors = ['r', 'g', 'b']

plt.bar(labels, values, color = colors, alpha = 0.5)
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]

plt.bar(labels, values)
plt.ylim(175, 195)
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]

plt.bar(labels, values, width = 0.5)
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]

plt.bar(labels, values, width = 0.3)
plt.xticks(rotation = 45) # x축의 이름 데이터 각도를 45도로 설정
plt.yticks(rotation = 45)
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]
ticks = ['1번 학생', '2번 학생', '3번 학생']

plt.bar(labels, values)
plt.xticks(labels, ticks, rotation = 45)
```
